````python
import requests
from bs4 import BeautifulSoup
import rasterio
from rasterio.windows import from_bounds

# Base URL
base_url = "http://files.isric.org/soilgrids/former/2017-03-10/data/"

# Get webpage
response = requests.get(base_url)
soup = BeautifulSoup(response.text, "html.parser")

print("Available TIFF files:\n")

unique_prefix = set()   # optional

for link in soup.find_all("a"):
    href = link.get("href")
    if href and href.endswith(".tif"):
        prefix = href[:5]   # commentout if no need
        if prefix not in unique_prefix:   # same as the above
            print(href)
            unique_prefix.add(prefix) # same
````

````python
# API
lat_min, lat_max = 28, 32
lon_min, lon_max = 77, 82
````

````python
base_url = "http://files.isric.org/soilgrids/former/2017-03-10/data/"

# Building filename
filename = f"{soil_parameter}_{statistic}_{depth}_250m_ll.tif" #or mannualy
url = base_url + filename
print("Downloading:", filename)

# Download file
response = requests.get(url, stream=True)
if response.status_code != 200:
    print("File not found. Check parameter names.")
else:
    with open(filename, "wb") as f:
        for chunk in response.iter_content(8192):
            f.write(chunk)
    print("Download complete.")

    # Clip
    with rasterio.open(filename) as src:

        # as file name ends with  "_ll", CRS is EPSG:4326
        bbox = (lon_min, lat_min, lon_max, lat_max)

        window = from_bounds(*bbox, transform=src.transform)
        clipped = src.read(1, window=window)

        profile = src.profile
        profile.update({
            "height": clipped.shape[0],
            "width": clipped.shape[1],
            "transform": src.window_transform(window)        })

        output_name = f"{soil_parameter}_{statistic}_{depth}_CLIPPED.tif"

        with rasterio.open(output_name, "w", **profile) as dst:
            dst.write(clipped, 1)

    print("Clipped file saved as:", output_name)
````
